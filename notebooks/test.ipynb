```python
from classy import Class
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(context='notebook', style='whitegrid', font_scale=1.5, rc={"figure.figsize":(8,5)})
```

```python
l_max = 2500
common_settings = {
    'output':'tCl, gwCl',
    'l_max_scalars':l_max,
    'write_warnings':'yes'}
M = Class()

M.set(common_settings)
M.compute()
cl = M.raw_cl(l_max)
```

```python
plt.figure()
plt.xlim([2,l_max])
plt.xlabel(r"$\ell$")
plt.ylabel(r"$\ell (\ell+1) C_l$")

ell = cl['ell']
factor = ell*(ell+1.)
plt.semilogx(ell,factor*cl['tt'],label=r'$C_\ell^\mathrm{TT}$')
# plt.semilogx(ell,factor*cl['gw'], '.',label=r'$C_\ell^\mathrm{GW}$')

plt.legend()
```

```python
prim = M.get_primordial()
print(prim.keys())

plt.figure()
plt.xlabel(r"$k , [1/\mathrm{Mpc}]$")
plt.ylabel(r"$P(k)$")

plt.plot(prim['k [1/Mpc]'], prim['P_scalar(k)'])

plt.xscale('log')
plt.yscale('log')

plt.legend()
```
